```python
import os
import pandas as pd
from dotenv import load_dotenv
from services.sheets_client import google_client_authentication


load_dotenv()
```

```python
spreadsheet_id = os.getenv("SHEET_ID")
google_credentials = os.getenv('GOOGLE_SHEETS_CREDENTIALS')

sheet = google_client_authentication(google_credentials, spreadsheet_id)
```

```python
raw_data = sheet.worksheet("Pedidos Processados")

v = raw_data.get_all_values()
df = pd.DataFrame(v)
```

```python
duplicated = df[df[2].duplicated(keep='first')]
print(duplicated)
```

```python
duplicated_values = df[2][df[2].duplicated(keep=False)]
print(duplicated_values)
```

```python
duplicated_rows = df[df[2].duplicated(keep=False)]
print(duplicated_rows)
```

```python
duplicados = df[df[2].duplicated(keep=False)]
print(duplicados)
```

```python
df_sem_duplicados = df.drop_duplicates(subset=[2], keep='first')
print(df_sem_duplicados)
```

```python
duplicados_restantes = df[df[2].duplicated(keep=False)]
print(duplicados_restantes)
```

```python
# Contar quantos duplicados existem na coluna 2
quantidade_duplicados = df[2].duplicated().sum()
print(f'Quantidade de duplicados: {quantidade_duplicados}')
```

```python
# Remover duplicatas com inplace=True
df.drop_duplicates(subset=[2], keep='first', inplace=True)

# Verificar se não há duplicados
quantidade_duplicados = df[2].duplicated().sum()
print(f'Quantidade de duplicados: {quantidade_duplicados}')
```

```python
duplicados_restantes = df[df[2].duplicated(keep=False)]
print(duplicados_restantes)
```
